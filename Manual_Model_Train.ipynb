```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(["age","cardio","Unnamed: 0",'smoke_alco'],axis=1)
y=data["cardio"]
```

## split data into train and test

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

## strndarize apply on input data

```python
scaler=StandardScaler()
x_train_scaler=scaler.fit_transform(x_train)
x_test_scaler = scaler.transform(x_test)
```

## convert output into probability between range 0 to 1,enabl binary classifcation

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
### sigmoid formula->z=XW+b
```

```python
def compute_cost(y,y_pred):
    m=len(y)
    #formula=−1/m∑[ylog(y^)+(1−y)log(1−y^)]
    cost=-(1/m)*np.sum(
        y*np.log(y_pred+1e-9)+(1 - y) * np.log(1 - y_pred + 1e-9) #log0 math error then use 1e-9
    )
    return cost
```

```python
def model_train(x,y,lr,itretaion):
    m, n = x_train_scaler.shape   # m->rows, n->columns
    w = np.zeros(n)               # for each column create weight
    b = 0 
    y=y.values
    for i in range(itretaion):
        z=np.dot(x,w)+b
        y_pred=sigmoid(z)
        
        dw = (1/m) * np.dot(x.T, (y_pred - y))
        db = (1/m) * np.sum(y_pred - y)

        w=w-lr*dw
        b=b-lr*db

        if i%100==0:
            cost=compute_cost(y,y_pred)
            # print(cost)
    return w, b
```

```python
w,b=model_train( 
    x_train_scaler,
    y_train, 
    lr=0.1,
    itretaion=20000 )
```

```python
def predict(x, w, b):
    probs = sigmoid(np.dot(x, w) + b)
    return (probs >= 0.5).astype(int)
```

```python
y_test_pred = predict(x_test_scaler, w, b)
accuracy = np.mean(y_test_pred == y_test)
print("Test Accuracy:", accuracy)
```
