```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cardio_train=pd.read_csv('Exploring_Dataset.csv')
cardio_train
```

```python
import sklearn
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
numeric_features = ['age','height','weight','ap_hi', 'ap_lo','age_in_year','bmi']
categorical_features = ['gender','cholesterol','gluc','smoke','alco','active','cardio','smoke_alco']
```

```python
# Create ColumnTransformer
preprocesser=ColumnTransformer(
    transformers=[
    ('num',StandardScaler(),numeric_features),
    ('cat',OneHotEncoder(handle_unknown='ignore'),categorical_features)]
)
preprocesser
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline=Pipeline(steps=[('preprocesser',preprocesser)])

from sklearn import set_config
set_config(display='diagram')
pipeline
```

```python
x=cardio_train[numeric_features + categorical_features]

x_processed=pipeline.fit_transform(x)

print(f"Original Shape : {x.shape}")
print(f"processed Shape :{x_processed}")
```

```python
new_columns=pipeline.named_steps['preprocesser'].get_feature_names_out()

df_processed=pd.DataFrame(x_processed,columns=new_columns)
df_processed.head()
```

```python
df_processed.to_csv('preprocess_cardio_train.csv')
```
